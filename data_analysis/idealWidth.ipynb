# Análisis de la optimización de carriles

Primero se define la función matemática que representa cada plano de cada uno de los patios. Tras tener dichas funciones, se analiza la distribución del ancho de los equipos, para seleccionar el ancho de los carriles. Teniendo un ancho definido para los carriles y las funciones, se posibilita la optimización de los planos.

## Obtener el ancho del carril

```python
# Import the equipments dataset
import pandas as pd
inventory = pd.read_excel("inputs/Equipos.xlsx")
```

```python
# Inspect the columns of the dataset
inventory.columns
```

```python
# Get the list of equipment width and length
widths = inventory['Ancho+FS (m)']
lengths = inventory['Largo+FS (m)']
```

```python
# Plot the Widths
import matplotlib.pyplot as plt

# Create the histogram plot
plt.hist(widths, bins=50, alpha=0.5)

# Add axis labels and a title
plt.xlabel('Equipment Widths')
plt.ylabel('Frequency')
plt.title('Equipment Width Distribution')

# Show the plot
plt.show()
```

```python
# Get  width stats
print(widths.describe())
```

De acuerdo con los resultados de las métricas del ancho de los equipos, usar 4.4m como ancho estandar de los carriles es una buena opción. Esto se debe a que los carriles deben ser de mínimo 3.22m, deben tener la capacidad de albergar hasta un ancho de 43.25m. Sin embargo, la mediana del ancho está en 8.42m y se ve una tendencia grande a tener equipos entre el valor mínimo y el valor de la mediana. Por lo que con dos carriles de 4.4m (8.8m), sería posible albergar los equipos de ancho más común. Teniendo en cuenta también el percentil del 75%, que es 13m, con 3 carriles de 4.4m (13.2m) se podrían guardar equipos de ese ancho. También, usando 10 carriles de 4.4m (44m) es posible albergar los equipos más grandes, que son de 43.25m.

```python
# Plot the Lenghts
import matplotlib.pyplot as plt

# Create the histogram plot
plt.hist(lengths, bins=30, alpha=0.5, color='red')

# Add axis labels and a title
plt.xlabel('Equipment Lengths')
plt.ylabel('Frequency')
plt.title('Equipment Length Distribution')

# Show the plot
plt.show()
```
